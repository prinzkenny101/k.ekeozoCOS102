```python
import tkinter as tk
from tkinter import filedialog as dialog, messagebox as msgbox
import csv

def check_employee(name, department):
    try:
        with open("GIG-logistics.csv", mode='r', encoding='utf-8') as file:
            reader = csv.DictReader(file)

            for row in reader:
                user_input = name.strip().upper()
                user_dept = department.strip().upper()
              
                csv_name = row['Name'].strip().upper()
                csv_dept = row['Department'].strip().upper()

                if (user_input == csv_name) and (user_dept == csv_dept):
                    return True
                    
    except FileNotFoundError:
        msgbox.showerror("Error", "CSV file not found")
        return False

def welcomeMessage(name, department):
    window = tk.Toplevel(root)
    window.title("GIG Logistics")
    window.geometry("900x500")

    label_1 = tk.Label(window, text=f"Welcome {name}\n", font=("Arial", 12, "bold"))
    label_1.pack()
    label_2 = tk.Label(window, text="This is GIG Logistics")
    label_2.pack()
    label_3 = tk.Label(window, text=f"\nOther members in {department}:", font=("Arial", 10, "underline"))
    label_3.pack()

    frame = tk.Frame(window)
    frame.pack(pady=10, fill="both", expand=True)

    scrollbar = tk.Scrollbar(frame)
    scrollbar.pack(side="right", fill="y")

    listbox = tk.Listbox(frame, yscrollcommand=scrollbar.set, width=40)
    listbox.pack(side="left", padx=10, fill="both", expand=True)
    scrollbar.config(command=listbox.yview)

    other_members = ""

    listbox.delete(0, tk.END)
    
    with open("GIG-logistics.csv", mode='r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            if row['Department'].strip().upper() == department.strip().upper():
                member_name = f"{row['Name']}"
                if name.strip().upper() not in member_name.upper():
                    listbox.insert(tk.END, member_name)

    list_label = tk.Label(window, text=other_members, justify="left")
    list_label.pack(pady=10)

def submit():
    name = name_entry.get()
    department = department_entry.get()

    if not name or not department:
        msgbox.showwarning("Input Error", "Please fill in both name and department.")
        return
        
    if check_employee(name, department):
        welcomeMessage(name, department)
    else:
        msgbox.showerror("Login Failed", "Employee does not exist or department is incorrect.")

def clear_fields():
    name_entry.delete(0, tk.END)
    department_entry.delete(0, tk.END)
```

```python
root = tk.Tk()
root.title("Login Page")
root.geometry("900x500")

information_title = tk.Label(root, text="Welcome to GIG Logistics", font=("Arial", 20, "bold"))
information_title.pack(pady=10) 
information_subtitle = tk.Label(root, text="Please verify your identity", font=("Helvetica", 12, "italic"))
information_subtitle.pack(pady=5)

name_label = tk.Label(root, text="Enter your Name:")
name_label.pack()
name_entry = tk.Entry(root)
name_entry.pack()

department_label = tk.Label(root, text="Enter your Department:")
department_label.pack(pady=(10, 0))
departments = ["Operations", "Finance", "Human Resources", "Technology", "Customer Service", "Sales", "Marketing"]
department_entry = tk.StringVar(root)
department_entry.set(departments[0])

department_menu = tk.OptionMenu(root, department_entry, *departments)
department_menu.pack(pady=5)

submit_button = tk.Button(root, text="Submit", command=submit)
submit_button.pack(pady=20)
clear_button = tk.Button(root, text="Clear Entries", command=clear_fields)
clear_button.pack(pady=5)

root.mainloop()
```

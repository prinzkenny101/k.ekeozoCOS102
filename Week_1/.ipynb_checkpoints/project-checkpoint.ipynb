```python
#Simple Interest Calculation

#collecting inputs
principal = abs(float(input("enter the principal:")))
rate = abs(float(input("enter the rate:")))
time = abs(float(input("enter the time:")))

#calculating amount
amount = principal * (1 + (rate / 100) * time)

#displaying amount
print("The simple interest amount is",  amount)
```

```python
#Compound Interest Calculation

#collecting inputs
principal = abs(float(input("enter the principal:")))
rate = abs(float(input("enter the rate:"))) / 100
time = abs(float(input("enter the time:")))
frequency = abs(float(input("enter the frequency:")))

#calculating amount
amount = principal * (1 + (rate / frequency)) ** (frequency * time)

#displaying amount
print(f"The compound interest amount is {amount:.2f}")
```

```python
#Annuity Plan Calculation

#collecting inputs
PMT = abs(float(input("enter the PMT:")))
rate = abs(float(input("enter the rate:"))) / 100
time = abs(float(input("enter the time:")))
frequency = abs(float(input("enter the frequency:")))

#calculating amount
amount = PMT * (((1 + (rate / frequency)) ** (frequency * time) - 1) / (rate / frequency))

#displaying amount
print(f"The annuity plan amount is {amount:.2f}")
```

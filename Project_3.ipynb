#Algorithm to put a list of students in a table

START

LOAD LIBRARIES
Import the pandas library as pd

DEFINE DATA (Girls)
Create a list girls_names
Create a list girls_ages
Create a list girls_height
Create a list girls_scores

DEFINE DATA (Boys)
Create a list boys_names
Create a list boys_ages 
Create a list boys_height
Create a list boys_scores

MERGE DATA
Combine all name lists into one all_names list
Combine all age lists into one all_ages list
Repeat for heights and scores

MAP TO DICTIONARY
Create a dictionary where"
    "Name" maps to all_names
    "Age" maps to all_ages
    "Height" maps to all_heights
    "Score" maps to all_scores

CONSTRUCT TABLE
    Pass the dictionary into the pandas.DataFrame() function

PRINT the resulting DataFrame

Arrange lists in a table

Sort the names alphabetically

STOP

```python
import pandas as pd

#Creating a list for the girls
girls_data = {'girls_names': ['Evelyn', 'Jessica', 'Somto', 'Edith', 'Liza', 'Madonna', 'Waje', 'Tola', 'Aisha', 'Latifa'],
              'girls_age': [17, 16, 17, 18, 16, 18, 17, 20, 19, 17],
              'girls_height': [5.5, 6.0, 5.4, 5.9, 5.6, 5.5, 6.1, 6.0, 5.7, 5.5],
              'girls_scores': [80, 85, 70, 60, 76, 66, 87, 95, 50, 49]}

#Creating a list for the boys
boys_data = {'boys_names': ['Chinedu', 'Liam', 'Wale', 'Gbenga', 'Abiola', 'Kola', 'Kunle', 'George', 'Thomas', 'Wesley'],
             'boys_age': [19, 16, 18, 17, 20, 19, 16, 18, 17, 19],
             'boys_height': [5.7, 5.9, 5.8, 6.1, 5.9, 5.5, 6.1, 5.4, 5.8, 5.7],
             'boys_scores': [74, 87, 75, 68, 66, 78, 87, 98, 54, 60]}

#Combining both lists
all_names = girls_data['girls_names'] + boys_data['boys_names']
all_ages = girls_data['girls_age'] + boys_data['boys_age']
all_heights = girls_data['girls_height'] + boys_data['boys_height']
all_scores = girls_data['girls_scores'] + boys_data['boys_scores']

#Creating a dictionary to mapp the combined list
final_data = {"Name": all_names, "Age": all_ages, "Height": all_heights, "Scores": all_scores}

#Passing the Dictionary into the pandas.Dataframe() function
student_table = pd.DataFrame(final_data)

#Sorting the data and giving the data proper index and printing the final result
s_table_sorted = student_table.sort_values(by="Name").reset_index(drop=True)
s_table_sorted.index = range(1, 21)
s_table_sorted.index.name = "S/N"
final_table = s_table_sorted.reset_index()
print(final_table.to_string(index=False))
```
